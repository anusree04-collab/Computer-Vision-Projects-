```python
import numpy as np
import cv2
```

```python
face_classifier = cv2.CascadeClassifier(r"C:\Users\K anusree\OneDrive\Desktop\FSDS\DEEP LEARNING\Face Detection\haarcascade_frontalface_default.xml")
```

```python
#Load the Image
img = cv2.imread(r"C:\Users\K anusree\OneDrive\Desktop\FSDS\DEEP LEARNING\Face Detection\Face.jpg")
```

```python
img
```

```python
#check if the image is loaded correctly or  not
if img is None:
    print("error: Image not Founded or cannot be loaded!")
    exit()
#convert the image into grayscale
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
```

```python
#Detect the faces in the image
faces = face_classifier.detectMultiScale(gray,1.3,5)
```

```python
#Check if faces are detected
if len(faces)==0:
    print("No Faces Found")
else:
    #Draw rectangles around the faces
    for (x, y, w, h) in faces:  # (x, y) is the top-left corner, and (w, h) is the width and height of the face
        cv2.rectangle(img, (x, y), (x + w, y + h), (127, 0, 255), 2)
        
```

```python
#Display the output
cv2.imshow("Face Detection",img)
cv2.waitKey(0)
```
