```python
import numpy as np
import cv2
```

```python
#Load the face and eye Haar cascade classifier
face_clasifier = cv2.CascadeClassifier(r"C:\Users\K anusree\OneDrive\Desktop\FSDS\DEEP LEARNING\Face Detection\haarcascade_frontalface_default.xml")
eye_classifier = cv2.CascadeClassifier(r"C:\Users\K anusree\OneDrive\Desktop\FSDS\DEEP LEARNING\Face Detection\haarcascade_eye.xml")
```

```python
img = cv2.imread(r"C:\Users\K anusree\OneDrive\Desktop\FSDS\DEEP LEARNING\Face Detection\Face.jpg")
```

```python
if img is None:
    print("Error: Image not found or cannot be loaded!")
    exit() 
```

```python
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
```

```python
faces = face_clasifier.detectMultiScale(gray,1.3,5)

# Check if faces are detected
if len(faces) == 0:
    print("No Face Found")
```

```python
for (x, y, w, h) in faces:
    # Draw a rectangle around the face
    cv2.rectangle(img, (x, y), (x + w, y + h), (127, 0, 255), 2)
```

```python
# Region of interest (ROI) for face
roi_gray = gray[y:y + h, x:x + w]
roi_color = img[y:y + h, x:x + w]
```

```python
# Detect eyes within the face region
eyes = eye_classifier.detectMultiScale(roi_gray)
```

```python
for (ex, ey, ew, eh) in eyes:
        # Draw rectangle around each detected eye
        cv2.rectangle(roi_color, (ex, ey), (ex + ew, ey + eh), (255, 255, 0), 2)
```

```python
# Display the output image with rectangles around faces and eyes
cv2.imshow('img', img)

# Wait for a key press before closing the window
cv2.waitKey(0)
```
